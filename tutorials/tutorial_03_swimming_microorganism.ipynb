```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def stokeslet(x, y, x0, y0):
    u = 1 / np.sqrt(np.square(x-x0)+np.square(y-y0))+np.square(x-x0)/(np.square(x-x0)+np.square(y-y0))**(3/2)
    v = (y-y0)*(x-x0)/(np.square(x-x0)+np.square(y-y0))**(3/2)
    return u,v
```

```python
nx, ny = 64, 64
x = np.linspace(-3, 3, nx)
y = np.linspace(-3, 3, ny)
X, Y = np.meshgrid(x, y)
```

```python
fig = plt.figure(figsize =(8, 8))
ax = fig.add_subplot(111)
U0, V0 = stokeslet(X, Y, 0, 0)
U1, V1 = stokeslet(X, Y, 1, 1)
U2, V2 = stokeslet(X, Y, 1, -1)
# Plotting the streamlines with
# proper color and arrow
ax.streamplot(x, y, U0-0.5*U1-0.5*U2, V0-0.5*V1-0.5*V2,
              linewidth = 1, cmap = plt.cm.inferno,
              density = 2, arrowstyle ='->',
              arrowsize = 1.5)
```
