# Diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def concentration(x, sourcepos, t):
    z = 1. / np.sqrt(4. * np.pi * t) * np.exp(-np.square(x-sourcepos) / (4. * t))
    return z
```

```python
def concentration_multiple(x, nbsources, t):
    z = concentration(x, 0, t)
    for i in range(1, nbsources):
        z += concentration(x, 2*i, t)
        z += concentration(x, -2*i, t)
    return z
```

```python
num_plots = 5
```

```python
colormap = plt.cm.gist_ncar
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.jet(np.linspace(0, 1, num_plots))))
x = np.linspace(-10,10,100)
for i in range(1, num_plots + 1):
    plt.plot(x, concentration(x,i*i))
```

```python
x = np.linspace(-10,10,100)
for i in range(1, num_plots + 1):
    plt.plot(x, concentration(x,i*i))
```

```python
x = np.linspace(0,2,100)
for i in range(1, num_plots + 1):
    plt.plot(x, concentration(x,1,0.1*i)+concentration(x,-1,0.1*i))
```

```python
x = np.linspace(0,2,100)
for i in range(1, num_plots + 1):
    plt.plot(x, concentration(x,1,0.1*i*i)-concentration(x,-1,0.1*i*i))
```

```python
x = np.linspace(-1,1,100)
for i in range(1, num_plots + 1):
    plt.plot(x, concentration_multiple(x,20,1*i*i))
```
